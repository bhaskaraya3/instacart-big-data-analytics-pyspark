```python
from pyspark.sql.functions import col, when, count

df.select([
    count(when(col(c).isNull(), c)).alias(c)
    for c in df.columns
]).show()
```

```python
df.withColumn("new_columns",col("col1") + col("col2"))
```

```python
from pyspark.sql.functions import when

df.withColumn(
    "category_flag",
    when(col("age") > 30, "Senior").otherwise("Young")
)
```

```python
df.withColumn(
    "age_group",
    when(col("age") < 18, "Teen")
    .when((col("age") >= 18) & (col("age") < 60), "Adult")
    .otherwise("Senior")
)
```

```python
from pyspark.sql.functions import lower, upper, length

df.withColumn("name_lower", lower(col("name")))
df.withColumn("len", length(col("name")))
```

```python
from pyspark.sql.functions import to_date, year, month

df.withColumn("date", to_date("date_col"))
df.withColumn("year", year("date"))
```

```python
from pyspark.sql.functions import concat_ws

df.withColumn("full_name", concat_ws(" ", "first", "last"))
```

```python
from pyspark.sql.functions import avg, sum, max, min

df.groupBy("category").agg(
    avg("price"),
    sum("sales"),
    max("sales")
).show()
```
